## Task 1
Write a program that accepts a 3x3 matrix and:
* displays the average values for each column,
* displays the maximum values for each column,
* displays the sum of elements on main diagonal,
* displays the bar charts of these averages and maximum values of columns (column number on X axis).

If the user provides incorrect data, the program should display appropriate message. 

> Hint: See `min`, `max` and `trace` functions of `ndarray`.


**Example:**

Input:

```
21  23  36
44  36  8
6   42  84
```

Output:
```
21   23   36 
44   36    8 
 6   42   84 
Average: 24.0, 34.0, 43.0
Maximum: 44, 42, 84
Sum of elements on main diagonal: 141
```




````python
# WRITE YOUR CODE BELOW
import numpy as np
import matplotlib.pyplot as plt
a = list(map(int, input('Write the 3x3 matrix in line (values seperated by spaces): ').split()))
input_matrix=np.array(a).reshape(3, 3)

maxi=np.amax(input_matrix, axis=0)
aver=np.average(input_matrix, axis=0)
sume=np.ndarray.trace(input_matrix)

print('Average: '+ str(aver))
print('Maximum: '+ str(maxi))
print('Sum of elements on main diagonal: '+ str(sume))


plt.subplot(121)
plt.bar(list(range(3)),aver)
plt.xticks(list(range(3)))
plt.xlabel("Column number")
plt.ylabel("Average")
plt.title("Average in each column")

plt.subplot(122)
plt.bar(list(range(3)), maxi)
plt.xticks(list(range(3)))
plt.xlabel("Column number")
plt.ylabel("Max")
plt.title("Max in each column")
````

## Task 2
Write a program that accepts a 5x5 matrix $X$ and applies a 2x2 average filter to this matrix. This
means that the output matrix $Y$ will be 4x4 size and each element of the output matrix $Y$ is
calculated as the average value of 4 elements of the input matrix $X$.
Calculation (should be optimized):
```
Y[0,0]=X[0,0]+X[0,1]+X[1,0]+X[1,1])/4=(10+20+20+10)/4=15; Y[0,1]=(20+30+10+40)/4=25;
Y[0,2]=(30+40+40+70)/4=45; Y[0,3]=(40+50+70+40)/4=50, ... ;Y[3,3]=(30+10+10+30)/4=20
```

**Example:**

Input:

<span style="color:red">10  20</span>  30  40  50<br />
<span style="color:red">20  10</span>  40  70  40<br />
30  20  10  30  40<br />
40  20  50  <span style="color:blue">30  10</span><br />
70  30  20  <span style="color:blue">10  30</span><br />


Output:


<span style="color:red">15.0</span>  25.0  45.0  50.0<br />
20.0  20.0  37.5  45.0<br />
27.5  25.0  30.0  27.5<br />
40.0  30.0  27.5  <span style="color:blue">20.0</span><br />


````python
# WRITE YOUR CODE BELOW
import numpy as np
import matplotlib.pyplot as plt
a = list(map(int, input('Write the 5x5 matrix in line (values seperated by spaces): ').split()))
X=np.array(a).reshape(5, 5)

print('Input:\n'+str(X))
Y=np.zeros((4,4))
for i in range(4):
    for j in range(4):
        arr=X[i:i+2,j:j+2]
        Y[i,j]=np.average(arr)
        
print('\nOutput:\n'+str(Y))
````
