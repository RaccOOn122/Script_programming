### Task 2

Find all palindromes in `english.txt` file. A palindrome is a string that read in reverse order gives the same string.

* Step 1: Put all words longer than 1 to set (uniqueness).
* Step 2: If word `s == s[::-1]`, then the word `s` is a palindrome.
* Step 3: All palindromes put into the set.
* Step 4: Display all palindromes.

Palindromes: `aaa`, `abba`, `abcdedcba`, `aba`, `xyzyx`.



````python
# >>> WRITE YOUR CODE HERE <<<
def is_palindrome(s):
    return s==s[::-1]

f=open("english.txt","r",encoding="latin1")
words=f.read()
words=words.split(" ")
words_longer={''}
for word in words:
    if len(word)>1:
        words_longer.add(word)
        

palindrome={''}
words_longer=list(words_longer)

for i in range(len(words_longer)):
    if is_palindrome(words_longer[i])==True:
        palindrome.add(words_longer[i])

print(palindrome)
````

### Task 3

Find all anagrams in `english.txt` file.

* Step 1: Put all words longer than 1 in the set (uniqueness).
* Step 2: Make a dictionary where the key consists of the sorted letters of a word, and the value is a list of all words in which sorted letters forms a key.

    For following words: `alga`, `pakt`, `gala`, `ptak`, `lata`, the dictionary should look like below:
```python
{
    'aagl': ['alga', 'gala'],
    'akpt': ['pakt', 'ptak']
}
```

* Step 3: Print all elements of the dictionary which contain more than one word.


````python
# >>> WRITE YOUR CODE HERE <<<
def anagrams(s):
    dictionary={}
    for i in s:
        key="".join(sorted(i))
        if key in dictionary:
            dictionary[key].append(i)
        else:
            dictionary[key]=[i]

    return dictionary
    
f=open("english.txt","r",encoding="latin1")
words=f.read()
words=words.split(" ")
words_longer={''}
for word in words:
    if len(word)>1:
        words_longer.add(word)
    
dict_anagrams=anagrams(words_longer)
dict_anagrams
for i in dict_anagrams.values():
    if len(i)>1:
        print('Anagrams: {}'.format(', '.join(i)))
````
