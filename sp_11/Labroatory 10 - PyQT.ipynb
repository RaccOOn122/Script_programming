### Task 2. Reverse Polish Notation

Create a function which takes the list with parameters in Reverse Polish Notation (RPN) and returns the result.

| Infix notation   | RPN |
| :----            | ---:      |
| 2 + 2            | 2 2 +     |
| (1 + 2) / 3      | 1 2 + 3 / |
| 1 + 2 / 3        | 1 2 3 / + |

````python
def RPN(li):
    stack=list()
    for i in li:
        stack.append(i)
    RPN_digit=list()
    RPN_operation=list()

    operations={
        '+':lambda num1, num2: num1+num2,
        '*':lambda num1, num2: num1*num2,
        '-':lambda num1, num2: num1-num2,
        '/':lambda num1, num2: num1/num2}

    for j in range(len(stack)):
        el=stack.pop()
        if el in operations.keys():
            RPN_operation.insert(0,el)
        else:
            RPN_digit.insert(0,el)
    return RPN_digit+RPN_operation
````
